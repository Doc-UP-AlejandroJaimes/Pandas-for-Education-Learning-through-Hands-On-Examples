### **Exercise 3: Saving and Loading CSV Files**  
**Objective:** Learn how to persist data using CSV format.  

##### **Task:**  
1. Save the DataFrame as `covid_colleges_backup.csv`.  
2. Reload the CSV file into a new DataFrame and confirm it matches the original.  
3. Load only the first 100 rows from the CSV file into a DataFrame.  

1. Save the DataFrame as `covid_colleges_backup.csv`.  

````python
file_bk = 'covid_colleges_backup.csv'
college_covid19_usa.to_csv(file_bk, index=False)
````

2. Reload the CSV file into a new DataFrame and confirm it matches the original.  

````python
college_covid19_bk = pd.read_csv(file_bk)
college_covid19_bk.equals(college_covid19_usa)
````

3. Load only the first 100 rows from the CSV file into a DataFrame.

````python
college_covid19_row100 = pd.read_csv(file_bk, nrows=100)
print(college_covid19_row100.shape)
college_covid19_row100.sample(n= 5, random_state=42)
````
